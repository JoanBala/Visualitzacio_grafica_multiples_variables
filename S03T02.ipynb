### Calculem la Mediana:

#### Amb la funció median() del mòdul 'statistics' de python

```python
round(propines['total_bill'].median(),2)    # Equival al valor de '50%' de 'describe()'
```

#### Fent el càlcul amb la operació matemàtica corresponent

La Mediana representa el valor de la variable de la posició central en un conjunt de dades ordenades.  
#Si la sèrie té un número parell de puntuacions, la Mediana és la mitja entre les dues puntuacions centrals ((n/2)+(n/2+1))/2. 

```python
Median_50 = round((propines['total_bill'].sort_values().iloc[122] + propines['total_bill'].sort_values().iloc[123]) / 2, 2)
Median_50
```

### Calculem la Mitja:

#### Amb la funció mean() del mòdul 'statistics' de python

```python
round(propines['total_bill'].mean(),2)    # Equival al valor del 'mean' de 'describe()'
```

#### Fent el càlcul amb la operació matemàtica corresponent

La mitja representa el promig. És una mesura de tendència central.  
#S'obté dividint la suma de totes les dades entre el total de dades.

```python
Mitja = round(propines['total_bill'].sum() / propines['total_bill'].count(), 2)
Mitja
```

```python
from IPython.display import display
pd.set_option('display.max_columns', None)

# Carreguem el DataSet amb el paquet Pandas...
pelis = pd.read_csv("movies_ansi.dat", sep="::", header=None, names=["ID", "Titol", "Categoria"])
pelis
```

```python
pelis = pelis.drop('ID', axis=1)
```

```python
pelis.head()
```

```python
pelis.info()
```

```python
pelis.describe()
```

```python
pelis
```

```python
#Extract from column 'Titol' the year in () and create a new column with this data
pelis['Any']= pelis['Titol'].str.extract(r'(\d{4})')
pelis.replace('\(\d{4}\)', '', regex=True, inplace=True)
pelis
```

```python
#Extract from column 'Categoria' the different types of movies

pelis['Genere']= pelis['Categoria'].str.split('|')
#pelis.drop('Categoria', axis=1, inplace=True)
pelis
```

```python
#Pelis per any (La historia del cine como espectáculo comenzó en París el 28 de diciembre de 1895.)
#'1600', '2001', '2010', '3000', '8230', '9000'
pelis_any = pelis.Any.value_counts().sort_index()
pelis_any.drop(pelis[pelis['Any'] < 1900].index, inplace=True)
```

```python
plt.figure(figsize =(16, 7))
#sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.lineplot(x=pelis_any.index, y=pelis_any.values)
```
